## **`Outliers removal techniques - IQR Method`**
 

**Here comes cool Fact for you!**

IQR is the first quartile subtracted from the third quartile; these quartiles can be clearly seen on a box plot on the data.

- Calculate IQR  and give a suitable threshold to remove the outliers and save this new dataframe into df2.

Let us help you to decide threshold: Outliers in this case are defined as the observations that are below (Q1 − 1.5x IQR) or above (Q3 + 1.5x IQR)

````python
import pandas as pd 
import numpy as np
Q1 = np.percentile(df["MSRP"], 25)
Q3 = np.percentile(df["MSRP"], 75)

# Calculate IQR
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("Interquartile Range (IQR):", IQR)
## define Q1 and Q2
# Q1 = 
# Q3 = 
#df2=IQR

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df2 = df["MSRP"][(df["MSRP"] < lower_bound) | (df["MSRP"] > upper_bound)]

df2
# 3️⃣ Filter out outliers
#df2 = MSRP[(MSRP >= lower_bound) & ("MSRP" <= upper_bound)]


# # define IQR (interquantile range) 
# IQR =    

# # define df2 after removing outliers
# df2 = 
df2
````

## `Heatmaps`


A heat map is a two-dimensional representation of information with the help of colors. Heat maps can help the user visualize simple or complex information

Using heatmaps plot the correlation between the features present in the dataset.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# 1) Create bins / categories
df['MSRP_bin'] = pd.qcut(df['MSRP'], q=8, duplicates='drop')
df['Cylinders_bin'] = pd.cut(df['Engine Cylinders'],
                             bins=[0,4,6,8,12],
                             labels=['<=4','5-6','7-8','>8'])

# 2) Pivot the data to create the matrix
pivot = df.pivot_table(index='Cylinders_bin',
                       columns='MSRP_bin',
                       values='Engine HP',
                       aggfunc='mean')

# 3) Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot,
            annot=True,
            fmt=".1f",
            cmap="YlOrRd",
            linewidths=0.5,
            linecolor='gray')
plt.title('Avg Engine Horsepower by Cylinders & MSRP bins')
plt.xlabel('MSRP (binned)')
plt.ylabel('Engine Cylinders (binned)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````
