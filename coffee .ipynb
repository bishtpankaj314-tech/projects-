```python
coffee['date']=pd.to_datetime(coffee['date'])
coffee['datetime']=pd.to_datetime(coffee['datetime'])
```

```python
coffee['month']=coffee['date'].dt.strftime('%m')
coffee['day']=coffee['date'].dt.strftime('%w')
coffee['hour']=coffee['datetime'].dt.strftime('%H')
coffee["year"]=coffee["date"].dt.strftime("%y")
```

```python
coffee
```

```python
revenue_data =coffee.groupby(['coffee_name']).sum(['money']).reset_index().sort_values(by='money',ascending=False)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=revenue_data,x="coffee_name",y="money")
plt.title("Sales by Coffee Catageory")
```

```python
monthly_sales =coffee.groupby(['coffee_name','month']).count()['date'].reset_index().rename(columns={'date':'count'}).pivot(index='month',columns='coffee_name',values='count').reset_index()
monthly_sales
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(data=monthly_sales)
plt.legend(loc='upper left')
plt.xticks(range(len(monthly_sales['month'])),monthly_sales['month'],size='small')
```

```python
weekday_sales =coffee.groupby(['day']).count()['date'].reset_index().rename(columns={'date':'count'})
weekday_sales
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=weekday_sales,x='day',y='count',color='steelblue')
plt.xticks(range(len(weekday_sales['day'])),['Sun','Mon','Tue','Wed','Thur','Fri','Sat'],size='small')
plt.title("Total no of sales day wise")
```

```python
daily_sales =coffee.groupby(['coffee_name','date']).count()['datetime'].reset_index().reset_index().rename(columns={'datetime':'count'}).pivot(index='date',columns='coffee_name',values='count').reset_index().fillna(0)
daily_sales
```

```python
hourly_sales =coffee.groupby(['hour']).count()['date'].reset_index().rename(columns={'date':'count'})
hourly_sales
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=hourly_sales,x='hour',y='count',color='steelblue')
#plt.xticks(range(len(hourly_sales['day'])),['Sun','Mon','Tue','Wed','Thur','Fri','Sat'],size='small')
plt.title("hour wise no of sales")
```

```python
hourly_sales_by_coffee =coffee.groupby(['hour','coffee_name']).count()['date'].reset_index().rename(columns={'date':'count'}).pivot(index='hour',columns='coffee_name',values='count').fillna(0).reset_index()
hourly_sales_by_coffee
```

```python

fig, axs = plt.subplots(2, 4, figsize=(20, 10))
axs = axs.flatten()
for i, column in enumerate(hourly_sales_by_coffee.columns[1:]):
# Skip the first column ('Index')
    axs[i].bar(hourly_sales_by_coffee['hour'],
    hourly_sales_by_coffee[column])
    axs[i].set_title(f'{column}')
    axs[i].set_xlabel('Hour')
#axs[i].set_ylabel('Sales')

plt.tight_layout()

# Show the plot
plt.show()
```
